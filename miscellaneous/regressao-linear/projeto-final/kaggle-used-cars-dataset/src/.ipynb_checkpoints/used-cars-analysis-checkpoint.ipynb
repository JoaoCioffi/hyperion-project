```python
# Assuming you have a DataFrame named df_train
columns_to_check = ['price', 'odometer']  # Replace with the names of your columns

# Define a function to remove outliers and NaN values from a specific column
def remove_outliers_and_nan(column):
    # Remove NaN values
    column = column.dropna()
    
    # Calculate quartiles and IQR for outlier detection
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1

    # Define a condition for outlier detection in this column
    outlier_condition = (column < (Q1 - 1.5 * IQR)) | (column > (Q3 + 1.5 * IQR))

    # Remove rows with outliers in this column
    return column[~outlier_condition]

# Loop through the columns and remove outliers and NaN values
for column_name in columns_to_check:
    df_train[column_name] = remove_outliers_and_nan(df_train[column_name])

# Now, df_train contains the columns with outliers and NaN values removed
df_no_outliers=df_train.copy()
```

## PrÃ©-modelagem

```python
numerics = ['int8', 'int16', 'int32', 'int64', 'float16', 'float32', 'float64']
categorical_columns = []
features = df_final.columns.values.tolist()
for col in features:
    if df_final[col].dtype in numerics: continue
    categorical_columns.append(col)
# Encoding categorical features
for col in categorical_columns:
    if col in df_final.columns:
        le = LabelEncoder()
        le.fit(list(df_final[col].astype(str).values))
        df_final[col] = le.transform(list(df_final[col].astype(str).values))
```

# Random Forest

```python
y = df_final.price.values

X = df_final.drop(['price'], axis=1)

from sklearn.model_selection import train_test_split
# Separar o conjunto em Treino/ Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV

model = GradientBoostingRegressor(loss='ls', learning_rate=0.1, n_estimators=100, subsample=1.0,
                                  criterion='friedman_mse', min_samples_split=2,
                                  min_samples_leaf=1,
                                  min_weight_fraction_leaf=0.0, max_depth=3,
                                  min_impurity_decrease=0.0, init=None, random_state=42,
                                  max_features=None,
                                  alpha=0.9, verbose=0, max_leaf_nodes=None, warm_start=False,
                                  validation_fraction=0.1, n_iter_no_change=None, tol=0.0001,
                                  ccp_alpha=0.0)

hyperparameters = dict(learning_rate=[0.05, 0.1, 0.2, 0.5, 0.9], loss=['squared_error', 'lad', 'huber'],
                       n_estimators=[20, 50, 100, 200, 300, 500, 700, 1000],
                       criterion=['friedman_mse', 'mse'], min_samples_split=[2, 4, 7, 10],
                       max_depth=[3, 5, 10, 15, 20, 30], min_samples_leaf=[1, 2, 3, 5, 8, 10],
                       min_impurity_decrease=[0, 0.2, 0.4, 0.6, 0.8],
                       max_leaf_nodes=[5, 10, 20, 30, 50, 100, 300])

randomized_search = RandomizedSearchCV(model, param_distributions=hyperparameters, random_state=42, n_iter=5,
                                       scoring=None, n_jobs=2, refit=True, cv=5, verbose=True,
                                       pre_dispatch='2*n_jobs', error_score='raise', return_train_score=True)
```

```python
hyperparameters_tuning = randomized_search.fit(X_train, y_train)
print('Best Parameters = {}'.format(hyperparameters_tuning.best_params_))
```
